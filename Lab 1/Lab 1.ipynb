**Сердюк, 474б. Вариант 8.** Ночь, месяц, лето, 5-этажный дом, часть окон светится, рядом с домом 2 ели.

**Шаг 1**. Создание пустого изображения с заданными размерами

```
from PIL import Image
from PIL import ImageDraw
import matplotlib.pyplot as plt

#задаем ширину и высоту изображения
width, height = 600,600

#функция вывода изображения
def draw_img(image):
  plt.figure(figsize=(8,8))
  plt.imshow(image)
  plt.show()

#создание нового изображения с заданными размерами
newImage = Image.new("RGB", (width, height)) 
#на изображении создаём объект для рисования
draw = ImageDraw.Draw(newImage)

#вывод полученного изображения
draw_img(newImage)
```

**Шаг 2.** Создание травы. Для этого поместим на нижнюю часть изображения прямоугольник зеленого цвета во всю ширину

```
#снег на полу - прямоугольник белого цвета с координатами: левая верхняя (0,0.65 высоты), правая нижняя (ширина, высота) 
draw.rectangle(((0, height * 0.65), (width, height)), '#3f9b0b')

#вывод полученного изображения
draw_img(newImage)
```

**Шаг 3.** Созданиие месяца. Для этого возьмём два эллипса разного цвета, перекрывающие друг друга


```
#луна основной эллипс жёлтого цвета
draw.ellipse(((0.05*width,0.05*height),(0.25*width,0.25*height)),'yellow')

#вывод полученного изображения
draw_img(newImage)
```

```
#перекрытие чёрным эллипсом для получения месяца
draw.ellipse(((0.025*width,0.025*height),(0.22*width,0.2*height)),'black')

#вывод полученного изображения
draw_img(newImage)
```

**Шаг 4.** Создание дома


Шаг 4.1. Создание основания дома. Для этого поместим на изображение прямоугольник корчневного цвета. Пусть это будет панельный дом с плоской крышей.

```
#дом
#основание дома - прямоугольник коричневого цвета
draw.rectangle(((width*2/6,height*1/12),(width,5/6*height)),'brown')

#вывод полученного изображения
draw_img(newImage)
```

Шаг 4.2. Создание окон. Для этого поместим на изображение прямоугольники цвета орхидеи

```
# рама - прямоугольник
# 1 этаж
draw.rectangle(((2.2/6*width,4/6*height),(3/6*width,4.6/6*height)),'orchid')
draw.rectangle(((4/6*width,4/6*height),(4.8/6*width,4.6/6*height)),'orchid')

# 2 этаж
draw.rectangle(((2.2/6*width,4/6.5*height),(3/6*width,4.6/8.9*height)),'orchid')
draw.rectangle(((4/6*width,4/6.5*height),(4.8/6*width,4.6/8.9*height)),'orchid')

# 3 этаж
draw.rectangle(((2.2/6*width,4/8.5*height),(3/6*width,4.6/12.6*height)),'orchid')
draw.rectangle(((4/6*width,4/8.5*height),(4.8/6*width,4.6/12.6*height)),'orchid')

# 4 этаж
draw.rectangle(((2.2/6*width,4/12.5*height),(3/6*width,4.6/21*height)),'orchid')
draw.rectangle(((4/6*width,4/12.5*height),(4.8/6*width,4.6/21*height)),'orchid')

# 5 этаж
draw.rectangle(((2.2/6*width,4/20*height),(3/6*width,4.6/45*height)),'orchid')
draw.rectangle(((4/6*width,4/20*height),(4.8/6*width,4.6/45*height)),'orchid')

# вывод полученного изображения

draw_img(newImage)
```

Шаг 4.3. Свет в окнах. Для этого поместим поверх прямоугольников цвета орхидеи, которые являются окнами, прямоугольники золотого цвета (свет включен) и синего цвета (свет выключен)

```
#свет в доме - прямоугольник
# 1
draw.rectangle(((2.25/6*width,4.05/6*height),(2.95/6*width,4.55/6*height)),'gold')
draw.rectangle(((4.05/6*width,4.05/6*height),(4.75/6*width,4.55/6*height)),'gold')

# 2
draw.rectangle(((2.25/6*width,3.95/6.5*height),(2.95/6*width,4.65/8.9*height)),'#000080')
draw.rectangle(((4.05/6*width,3.95/6.5*height),(4.75/6*width,4.65/8.9*height)),'gold')

# 3
draw.rectangle(((2.25/6*width,3.95/8.5*height),(2.95/6*width,4.7/12.6*height)),'gold')
draw.rectangle(((4.05/6*width,3.95/8.5*height),(4.75/6*width,4.7/12.6*height)),'gold')

# 4
draw.rectangle(((2.25/6*width,3.95/12.5*height),(2.95/6*width,4.7/21*height)),'gold')
draw.rectangle(((4.05/6*width,3.95/12.5*height),(4.75/6*width,4.7/21*height)),'#000080')

# 5
draw.rectangle(((2.25/6*width,3.9/20*height),(2.95/6*width,4.75/45*height)),'gold')
draw.rectangle(((4.05/6*width,3.9/20*height),(4.75/6*width,4.75/45*height)),'#000080')

#вывод полученного изображения
draw_img(newImage)
```

Шаг 4.4. Створки окон. Для этого поместим на изображение прямоугольники цвета орхидеи посреди прямоугольника окна

```
#створки - прямоугольник
#1
draw.rectangle(((2.575/6*width,4/6*height),(2.625/6*width,4.6/6*height)),'orchid')
draw.rectangle(((4.375/6*width,4/6*height),(4.425/6*width,4.6/6*height)),'orchid')

#2
draw.rectangle(((2.575/6*width,4/6.5*height),(2.625/6*width,4.6/8.9*height)),'orchid')
draw.rectangle(((4.375/6*width,4/6.5*height),(4.425/6*width,4.6/8.9*height)),'orchid')

#3
draw.rectangle(((2.575/6*width,4/8.5*height),(2.625/6*width,4.6/12.6*height)),'orchid')
draw.rectangle(((4.375/6*width,4/8.5*height),(4.425/6*width,4.6/12.6*height)),'orchid')

#4
draw.rectangle(((2.575/6*width,4/12.5*height),(2.625/6*width,4.6/21*height)),'orchid')
draw.rectangle(((4.375/6*width,4/12.5*height),(4.425/6*width,4.6/21*height)),'orchid')

#5
draw.rectangle(((2.575/6*width,4/20*height),(2.625/6*width,4.6/45*height)),'orchid')
draw.rectangle(((4.375/6*width,4/20*height),(4.425/6*width,4.6/45*height)),'orchid')

#вывод полученного изображения
draw_img(newImage)
```

Шаг 5. Создание елей. Для этого используем прямоугольники как основание для елей (цвета оттенка красного) и выстраиваем полигоны (которым зададим зеленый цвет), чтобы из треугольников сделать что-то похожее на ель

```
# ель №1
draw.rectangle(((1.35/2 * width, height * 0.85), (width // 1.5 + width // 2 * 0.05, height * 0.9)), '#935a52')

draw.polygon(((int(0.60 * width), int(height * 0.85)),
(int(0.77 * width), int(height * 0.85)),
(int(0.68 * width), int(height * 0.75)),
), '#1e6428')

draw.polygon(((int(0.62 * width), int(height * 0.78)),
(int(0.75 * width), int(height * 0.78)),
(int(0.68 * width), int(height * 0.68)),
), '#1e6428')

draw.polygon(((int(0.63 * width), int(height * 0.72)),
(int(0.74 * width), int(height * 0.72)),
(int(0.68 * width), int(height * 0.63)),
), '#1e6428')

# ель №2
draw.rectangle(((0.5/2 * width, height * 0.85), (width // 2 * 0.55, height * 0.9)), '#935a52')

draw.polygon(((int(0.18 * width), int(height * 0.85)),
(int(0.35 * width), int(height * 0.85)),
(int(0.26 * width), int(height * 0.75)),
), '#1e6428')

draw.polygon(((int(0.195 * width), int(height * 0.78)),
(int(0.33 * width), int(height * 0.78)),
(int(0.26 * width), int(height * 0.68)),
), '#1e6428')

draw.polygon(((int(0.205 * width), int(height * 0.72)),
(int(0.315 * width), int(height * 0.72)),
(int(0.26 * width), int(height * 0.63)),
), '#1e6428')

#вывод полученного изображения
draw_img(newImage)
```
